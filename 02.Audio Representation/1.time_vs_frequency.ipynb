```python
%matplotlib inline
import numpy, scipy, matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
```

## 1. Waveforms and the Time Domain
Let's listen to a file (if you get an error using librosa.load, you may need to *install ffmpeg*):

```python
x, sr = librosa.load("./c_strum.wav")
print(x.shape)
print(sr)
ipd.Audio(x, rate=sr)
```

Waveform:

```python
plt.figure(figsize=(15, 5))
librosa.display.waveplot(x, sr, alpha=0.8)
```

Zoom in:

```python
plt.figure(figsize=(15, 5))
plt.scatter(range(250), x[1000:1250], marker='.', alpha=0.8)
```

## 2. The frequency domain and Fourier Transform
Compute a Fourier transform

```python
X = scipy.fft(x) # Fourier transform
X_mag = numpy.absolute(X) # magnitude
f = numpy.linspace(0, sr, len(X_mag)) - sr/2# frequency variable
```

Plot the spectrum 

```python
plt.figure(figsize=(13, 5))
plt.plot(f, X_mag) # magnitude spectrum
plt.xlabel('Frequency (Hz)')
```
