```python
import numpy, scipy, matplotlib.pyplot as plt, sklearn, librosa, librosa.display, urllib, IPython.display
```

## 1. Mel Frequency Cepstral Coefficients (MFCCs)

Let's load and plot an audio file:

```python
x, sr = librosa.load('./simple_loop.wav')
plt.figure(figsize=(15, 5))
librosa.display.waveplot(x, sr=sr)
print(x.shape)
print(sr)
IPython.display.Audio(x, rate=sr)
```

Compute MFCCs across the audio signal. In this case, `mfcc` computed 20 MFCCs over 97 frames.

```python
mfccs = librosa.feature.mfcc(x, sr=sr)
print(mfccs.shape)
```

Display the MFCCs:

```python
plt.figure(figsize=(15, 5))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
```

Normalise to zero mean and unit variance:

```python
mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
plt.figure(figsize=(15, 5))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
```
